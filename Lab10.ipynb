```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
```

```python
x = np.array([1.1, 1.4, 1.9, 2.5, 2.7])
y = np.array([2.91, 3.64, 4.55, 2.57, 0.24])

cs = CubicSpline(x, y)

xs = np.linspace(1.1, 2.7, 200)
ys = cs(xs)

plt.plot(x, y, 'o', label='Задані точки')
plt.plot(xs, ys, label='Кубічний сплайн')
plt.legend()
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Апроксимація кубічним сплайном")
plt.show()
```
