```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-x) + x**2

x = np.arange(0.1, 1.1, 0.1)
y = f(x)

coef_lin = np.polyfit(x, y, 1)
y_lin = np.polyval(coef_lin, x)

coef_par = np.polyfit(x, y, 2)
y_par = np.polyval(coef_par, x)

plt.plot(x, y, 'o', label='Точки даних')
plt.plot(x, y, label='f(x)')
plt.plot(x, y_lin, label='Лінійна МНК')
plt.plot(x, y_par, label='Параболічна МНК')

plt.legend()
plt.grid(True)
plt.show()
```
