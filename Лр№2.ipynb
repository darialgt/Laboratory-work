```python
import numpy as np

def f(x):
    return 2*x**4 - 2*x**3 - 4*x**2 + 6*x + 5

a, b = -10, 10
step = 0.5

roots_intervals = []
x = a
while x < b:
    if f(x) * f(x + step) < 0:
        roots_intervals.append((x, x + step))
    x += step

print("Інтервали, де є корені:")
for i, r in enumerate(roots_intervals, 1):
    print(f"{i}) [{r[0]}, {r[1]}]")
```

```python
import numpy as np

def f(x):
    return 2*x**4 - 2*x**3 - 4*x**2 + 6*x + 5

def bisection(f, a, b, eps=1e-4):
    if f(a)*f(b) > 0:
        raise ValueError("На відрізку немає кореня або парна кількість.")
    while (b - a)/2 > eps:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        if f(a)*f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

root = bisection(f, -1.5, -1.0)
print("Корінь методом бісекції:", root)
```

```python
import numpy as np

def f(x):
    return 2*x**4 - 2*x**3 - 4*x**2 + 6*x + 5

def chord(f, a, b, eps=1e-4):
    if f(a)*f(b) > 0:
        raise ValueError("На відрізку немає кореня або парна кількість.")
    x0, x1 = a, b
    while abs(x1 - x0) > eps:
        x2 = x1 - f(x1)*(x1 - x0)/(f(x1)-f(x0))
        x0, x1 = x1, x2
    return x1

root = chord(f, -1.5, -1.0)
print("Корінь методом хорд:", root)
```
