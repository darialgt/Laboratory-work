1. Обчислення інтеграла методом прямокутників

```python
from scipy import integrate
import numpy as np

a, b, n = 3.2, 4, 10
h = (b - a) / n  # крок

def f1(x):
    return 1 / np.sqrt(0.5 * x + 1)

# Метод лівих прямокутників
def left_rec(f1, a, b):
    s = 0
    for i in range(n):
        s += f1(a + i * h)
    return s * h

print("Left rectangle:", round(left_rec(f1, a, b), 5))

# Метод правих прямокутників
def right_rec(f1, a, b):
    s = 0
    for i in range(1, n + 1):
        s += f1(a + i * h)
    return s * h

print("Right rectangle:", round(right_rec(f1, a, b), 5))

# Метод середніх прямокутників
def aver_rec(f1, a, b):
    s = 0
    for i in range(n):
        s += f1(a + (i + 0.5) * h)
    return s * h

print("Average rectangle:", round(aver_rec(f1, a, b), 5))

# Перевірка SciPy
v, err = integrate.quad(f1, a, b)
print("SciPy quad:", round(v, 5))
```

2. Обчислення інтеграла методом Симпсона

```python
from scipy import integrate
import numpy as np

# Функція
def f(x):
    return x**2 * np.cos(x)

# Межі інтегрування та кількість розбиттів
a, b, n = 0.6, 1.4, 8

h = (b - a) / n

# Метод Сімпсона
def simpson_rule(f, a, b, n):
    integr = f(a) + f(b)
    for i in range(1, n):
        x = a + i * h
        if i % 2 == 0:
            integr += 2 * f(x)
        else:
            integr += 4 * f(x)
    integr *= h / 3
    return integr

# Результат
print("Simpson method:", round(simpson_rule(f, a, b, n), 4))

# Перевірка SciPy
v, err = integrate.quad(f, a, b)
print("Check:", round(v, 4))
```

3. Обчислення інтегралу методом трапецій

```python
from scipy import integrate
import numpy as np

# Функція
def f(x):
    return 1 / np.sqrt(3*x**2 + 1)

# Межі інтегрування та кількість розбиттів
a, b, n = 1.4, 2.2, 20

# Метод трапецій
def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    s = 0.5 * (f(a) + f(b))
    x = a
    for i in range(1, n):
        x += h
        s += f(x)
    return h * s

# Обчислення
integral = trapezoidal_rule(f, a, b, n)
print("Trapetzia method result:", round(integral, 5))

# Перевірка SciPy
v, err = integrate.quad(f, a, b)
print("Check with SciPy quad method:", round(v, 5))
```
